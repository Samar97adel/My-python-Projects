## 2) Why type casting matters

Even if the user types a number like `25`, Python receives it as text.

That means this:

```python
Age =int(input('Enter your age '))  # =>'27' as a string  =>  27 as an int
# Age =27
print(Age+5)
```

If you want to use the input as a number, you need **type casting**.

Common casting functions:
- `int()` → convert to integer
- `float()` → convert to decimal number
- `str()` → convert to text

```python

# Converting input to an integer
Age =int(input('Enter your age '))  # =>'27' as a string  =>  27 as an int
# Age =27
print(Age+5)
```

```python

# Converting input to a float
price= float(input('Enter the price of the product '))  #100.0
price_discount=price -20 #80.0
print('the price is',price ) #80
print('the price after the discount is ',price_discount)
print(type(price))
```

## 3) Common beginner mistake with `input()`

Look at this example:

```python

# Take two numbers from the user and print their addition , subtraction , multiplication and division
number1 = int(input('enter the first number '))
number2 = int(input('enter the second number '))
add=number1+number2
print('the addition of the two numbers is',add)
sub=number1-number2
print('the subtract of the two numbers is',sub)
mult=number1*number2
print('the multiplication of the two numbers is',mult)
div=number1/number2
print('the division of the two numbers is',div)
```

### What happens?
If the user enters:
- `5`
- `7`

The result may become:

`57`

Why? Because both values are **strings**, and string `+` means **concatenation** (joining text).

Let's fix it.

```python
# wrong answer
x = input("Enter first number: ")
y = input("Enter second number: ")

print("Correct numeric result =", x + y)
```

```python
# right answer
x = int(input("Enter first number: "))
y = int(input("Enter second number: "))

print("Correct numeric result =", x + y)
```

## 5) Introduction to `if` conditions

Programs often need to make decisions.

Examples:
- If score is high → print "Pass"
- If age is 18 or more → print "Adult"
- If number is even → print "Even"

The basic syntax is:

```python
# compare variable number by 5 and print if the var is bigger
num=10
if num > 5 :
    print('this number is Greater than 5 ')
    
else :    # لو الشرط ال ف ifمتحققش 
    print('this number is smaller than 5 ')
```

### Important syntax rule
Python uses **indentation** to show which lines belong to the `if` block.

## 6) `if` with comparison operators

Common comparison operators:
- `==` equal to
- `!=` not equal to
- `>` greater than
- `<` less than
- `>=` greater than or equal to
- `<=` less than or equal to

```python
# take from the user 2 numbers 
#if adding = 50 => add ==50
#if adding >50  => add > 50
#if adding >50  => add < 50
x= int(input('enter first number'))
y= int(input('enter second number'))

addition=x+y
if addition == 50 :
    print('the sum of the two numbers is 50')
elif addition >50 :
    print('the sum of the two numbers is greater than 50 , the number is' , addition)   
else:  #  number <50
    print('the sum of the two numbers is smaller than 50  , the number is' , addition)   

```

## 7) `if` and `else`

Use `else` when you want one result if the condition is true, and another result if it is false.

```python
# take from user txt2 => if txt1 =txt2 'same word' x => different words 
txt1=input('enter your text' )
txt2 =input('enter your text' )

if txt1 == txt2 :
    print('Same Word')
else:
    print('Different word')    
```

## 13) Common mistakes beginners make

### Mistake 1: Forgetting type casting
Trying to compare a string with a number may cause errors or wrong logic.

### Mistake 2: Using `=` instead of `==`
- `=` means assignment
- `==` means comparison

### Mistake 3: Wrong indentation
Python depends on indentation.

### Mistake 4: Expecting `input()` to return numbers automatically
It returns text unless you cast it.

```python

# Wrong:
# age = input("Enter age: ")
# if age >= 18:
#     print("Adult")

# Correct:
age = int(input("Enter age: "))
if age >= 18:
    print("Adult")
```
