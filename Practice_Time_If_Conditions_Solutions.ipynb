

## Dictionary Illustration

A **dictionary** stores data in **key : value** pairs.

```python
student = {
    "name": "Ali",
    "age": 20,
    "grade": "A"
}
```

You can imagine it like this:

| Key | Value |
|---|---|
| `"name"` | `"Ali"` |
| `"age"` | `20` |
| `"grade"` | `"A"` |

### Visual idea
```text
student
│
├── "name"  → "Ali"
├── "age"   → 20
└── "grade" → "A"
```

### Common dictionary operations
```python
student["name"]          # get a value
student["age"] = 21      # update a value
student["city"] = "Sana'a"   # add a new key

```

## Loops

````python
# 1) Even and Odd Separator
numbers=[1,2,3,4,5,6]
print ('even numbers are')
for i in numbers :
    if i %2==0:
        print(i)
    
print('odd numbers are')
for i in numbers :
    if i %2 !=0:
        print(i)
````

````python
# 2) Count Positive, Negative, and Zero Numbers
nums=[-1,0,1,2,3,-6,-5,0,0]

count_zeros=0
count_pos=0
count_neg=0

for num in nums :
    if num>0:
        count_pos+=1
    elif num==0 :
        count_zeros+=1   
    else:
        count_neg+=1   

print('Count of Zeros =',count_zeros)       
print('Count of positives =',count_pos)          
print('Count of negatives =',count_neg)             
````

## Functions

````python
# 1) Even or Odd Counter
def count_even_odd(numbers):
    even_count = 0
    odd_count = 0

    for num in numbers:
        if num % 2 == 0:
            even_count += 1
        else:
            odd_count += 1

    return even_count, odd_count

nums = [1, 2, 3, 4, 5, 6]
even_count, odd_count = count_even_odd(nums)
print("Even:", even_count)
print("Odd:", odd_count)
````

````python
# 2) Lucky Number Checker
def check_lucky(numbers):  
    for num in numbers :#  47  '47' => '7'
        if num %7==0    and  str(num).endswith('7'):
            print (num ,'is a lucky number ')
        else:
            print (num ,'is not lucky number ')

nums=[7,67,77,10,147]
check_lucky(nums)
````

````python
# 3) Word Length Sorter
def word_length(words):
    for word in words :
        if len(word)>5:
            print(word ,'is long one')
        else:
            print(word ,'is short')    
words =['cat','pyrhon','ai','datascience']            
word_length(words)
````

````python
# 4) Find Largest Number (Using Loop + If)
def find_max(nums):
    maximum = nums[0]

    for num in nums:
        if num > maximum:
            maximum = num

    return maximum

nums = [12, 45, 7, 89, 34]
print("Largest number:", find_max(nums))
````

````python
# 5) Number Range Classifier (FizzBuzz)
def classify_numbers(n):
    for i in range(1, n + 1):
        if i % 3 == 0 and i % 5 == 0:
            print("FizzBuzz")
        elif i % 3 == 0:
            print("Fizz")
        elif i % 5 == 0:
            print("Buzz")
        else:
            print(i)

classify_numbers(20)
````

# Mini System 1: Library Book Borrowing System

## Task Description
Create a simple library system using OOP.

### Requirements
1. Create a class `Book` with:
   - `title`
   - `author`
   - `is_borrowed`

2. Create methods inside `Book`:
   - `borrow_book()` → marks the book as borrowed if available
   - `return_book()` → marks the book as returned
   - `display_info()` → prints book details

3. Create at least 2 book objects.

4. Test the system by:

   - borrowing a book
   - trying to borrow it again
   - returning it

## Goal
Practice:

- instance methods
- simple object state changes

````python
class book :
    def __init__(self, title, author):
        self.title=title
        self.author=author
        self.is_borrowed=False
    def borrow_book(self):
        if self.is_borrowed ==True:
            print('this book is borrowed') 
        else:
            print('the book has been borrwed successfully')   
            self.is_borrowed=True
    def return_book(self) :
        if self.is_borrowed :
            self.is_borrowed =False
            print('book is returned')
        else:
            print('the book was not borrowed')    
    def display_info(self):
        print(f'Title :{self.title}')
        print(f'Author :{self.author}')


book1=book('xyz', 'abc')
book1.borrow_book()
book1.borrow_book()

book1.return_book()        
````

# Mini System 2: Student Course Registration System

## Task Description
Create a course registration system using OOP.

### Requirements
1. Create a class `Course` with:
   - `course_name`
   - `instructor`
   - `capacity`
   - a list of registered students

2. Add methods inside `Course`:
   - `register_student(student_name)` → adds a student if there is space
   - `remove_student(student_name)` → removes a student if found
   - `display_course_info()` → prints course details and registered students

3. Create a class `Student` with:
   - `name`
   - `student_id`

4. Add a method in `Student`:
   - `display_student_info()`

5. Create at least:
   - 1 course
   - 3 students

6. Test the system by:
   - registering students
   - trying to register more than capacity
   - removing a student
   - displaying final course information



````python
class Student :
    def __init__(self,name,id):
        self.name=name
        self.id=id
    def display_info(self):
        print(f'student name :{self.name} , student id :{self.id}')

class Course:
    def __init__(self, course_name ,instructor,capacity):
        self.course_name=course_name
        self.instructor=instructor
        self.capacity=capacity
        self.registered_students=[]
    def register_student(self,student) :
        if student not in self.registered_students:
            if len(self.registered_students)  < self.capacity:
                self.registered_students.append(student)
            else:
                print(f'{student.name} you are already in the course')    
        else:
            print('Can not register , the capacity is full')       
    def remove_student(self,student):
            if student  in self.registered_students:
                self.registered_students.remove(student)
                print(f'{student.name} is removed')
            else:
                print(f'{student.name}is not in the course ')    
    def display_course_info(self):
        print(f'Course :{self.course_name}')            
        print(f'instructor :{self.instructor}')            
        print(f'Capacity :{self.capacity}')            
        print(f'Registered Students :{len(self.registered_students)}')  

std1=Student('Ali','1')
std2=Student('Ahmed','2')
std3=Student('Khalid','3')

course=Course('Data Science','Dina',2)

std1.display_info()

course.register_student(std1)
course.register_student(std2)
course.register_student(std3)

course.display_course_info()
course.remove_student(std1)
course.display_course_info()
````
