# 3️⃣ What is an Object?

An **object** is an actual item created from a class.

So:

- `Student` → class

````python

class Student:
    pass

student1 = Student()
student2 = Student()

print(student1)
print(student2)
````

## Important Note

Even if two objects come from the same class, they are still **different objects**.

That means each object can store its own data.


# 4️⃣ Attributes

Attributes are variables stored inside an object.

For a student, attributes may be:

- name
- age
- grade


````python

class Student:
    def __init__(self, name, age, grade):
        self.name = name
        self.age = age
        self.grade = grade

student1 = Student("Ali", 20, "A")
student2 = Student("Sara", 22, "B")

print(student1.name)
print(student2.grade)
````

# 6️⃣ Methods

Methods are functions inside a class.

They describe what the object can do.


````python

class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def introduce(self):
        print("Hello, my name is", self.name)

    def birthday(self):
        self.age += 1
        print(self.name, "is now", self.age)

student1 = Student("Mona", 19)

student1.introduce()
student1.birthday()
````

# 7️⃣ Practice 1

Create a class called `Car`.

It should have:

- `brand`
- `color`

And a method called:

- `show_info()`


````python
class BankAccount:

    def __init__(self,name, balance):
        # private variable (protected from direct access)
        self.__balance = balance
        self.name=name

    # allowed method to deposit money
    def deposit(self, amount):
        self.__balance += amount

    # allowed method to withdraw money
    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
        else:
            print("Not enough balance")

    # allowed method to read balance
    def get_balance(self):
        return self.__balance


# create account
account = BankAccount('Dina',1000)

account.name='ahmed'
account.__balance=100
print(account.name)
# correct way to modify data
account.deposit(500)
print("Balance after deposit:", account.get_balance())

account.withdraw(200)
print("Balance after withdraw:", account.get_balance())


# wrong way: trying to access private variable directly
#print(account.__balance)   # this will raise an error
````

````python

class Car:
    def __init__(self, brand, color):
        self.brand = brand
        self.color = color

    def show_info(self):
        print("Brand:", self.brand)
        print("Color:", self.color)

car1 = Car("Toyota", "Red")
car1.show_info()
````

# 8️⃣ Encapsulation

Encapsulation means we **protect data** and control how it is changed.

Instead of letting anyone change values in any way, we use methods with rules.


## Illustration: Protected Box

<div style="border:2px solid black; border-radius:12px; padding:18px; background:#fafafa;">
  <div style="font-size:18px; font-weight:bold;">BankAccount</div>
  <div style="margin-top:10px;">💰 balance</div>
  <div style="margin-top:10px;">🔒 Do not change carelessly</div>
  <div style="margin-top:10px;">Use: deposit() / withdraw()</div>
</div>


````python

class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self._balance = balance   # protected by convention

    def deposit(self, amount):
        if amount > 0:
            self._balance += amount
            print("Deposit successful")
        else:
            print("Invalid deposit amount")

    def withdraw(self, amount):
        if amount <= self._balance:
            self._balance -= amount
            print("Withdrawal successful")
        else:
            print("Not enough balance")

    def show_balance(self):
        print("Current balance =", self._balance)

account1 = BankAccount("Omar", 1000)
account1.deposit(300)
account1.withdraw(200)
account1.show_balance()
````

### Why is this better?

Because now we can apply rules such as:

- no negative deposit
- no withdrawal larger than balance

So the object becomes safer and smarter.

# 1️⃣1️⃣ Abstraction

**Abstraction** means we focus on **what an object should do** and hide the unnecessary details of **how it does it**.

In Python, we often use abstraction with **abstract classes**.

An abstract class is like a rule sheet:
- it defines required methods
- child classes must complete them
- we usually do not create objects directly from it


## Real-Life Idea

Think of a remote control.

You press **Power On**.  
You care that the TV turns on.  
You do **not** need to know all the internal electrical steps.

That is abstraction.


## Why use abstraction?

- It makes code more organized
- It forces child classes to follow a clear structure
- It is useful when different classes must provide the same method
