# Task 5: Write a function max_of_three(a,b,c) that returns the largest of 3 numbers
# (Use only if-elif-else, no built-in max)

```python
def max_of_three(a,b,c):
    if a>b and a>c:
        print(f'the largest number is {a}')
    elif b>a and b>c:
        print(f'the largest number is {b}')
    else:
        print(f'the largest number is {c}')
    return a,b,c

max_of_three(3,4,8)
```

# Task 7: Write a function reverse_string(text) that returns the reversed string
# Example: reverse_string("hello") → "olleh"

```python
def reverse_string(text):
    return text[::-1]

reverse_string('hello')
```

# Task 8: Write a function count_vowels(word) that returns number of vowels (a,e,i,o,u)
# Example: count_vowels("education") → 5


```python
def count_vowels(word):
    vowels='aeiouAEIOU'
    count=0
    for char in word:
        if char in vowels:
            count+=1
    return count

count_vowels('Education')
```

# Task 9: Create a list numbers = [1, 2, 3, 4, 5, 6, 7, 8]

# Task 10: Use filter() + lambda to get only even numbers from numbers = [10, 15, 20, 23, 30, 33]


```python
numbers=[10,15,20,23,30,33]
even_numbers=list(filter(lambda x:x%2==0,numbers))
print(even_numbers)
```

# Task 13: Write a function is_palindrome(word) → returns True if word reads same backwards
# Example: "radar" → True, "hello" → False


```python
is_palindrome=lambda word:word.lower()==word.lower()[::-1]
print(is_palindrome("Radar"))
print(is_palindrome("hello"))
```

```python
def is_palindrome(word):
    word=word.lower()
    return word==word[::-1]

print(is_palindrome("Radar")) 
print(is_palindrome("hello"))
```

# Use from functools import reduce
# Use reduce() + lambda to find the PRODUCT of all numbers in list [2, 3, 4, 5]
# Expected result: 120


```python
from functools import reduce
numbers=[2,3,4,5]
product= reduce(lambda x,y : x*y, numbers)
print(product)
```
