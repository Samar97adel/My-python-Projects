```python
class person:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def introduce(self):
        print('i am ',self.name)
        print('my age now is ',self.age)

class teacher(person):
    def teaching(self):
        print(self.name, 'is teaching')

t1=teacher('Ahmed',34)
t1.teaching()
```

```python
class Animal():
    def __init__(self,name,age,):
        self.name=name
        self.age=age
    def make_sound(self):
        print(self.name,'is making sound')
        
class cat(Animal):
    def make_sound(self):
        print(self.name,'is meowing')
class dog(Animal):
    def make_sound(self):
        print(self.name,' is barking')

c1=cat('bossycat',2)
c1.make_sound()
d1=dog('bobby',3)
d1.make_sound()
```

```python
class Vehicle:
    def __init__(self, brand,speed):
        self.brand=brand
        self.speed=speed
    def move(self):
        print(f'The vehicle {self.brand} moves with speed {self.speed}')

class car(Vehicle):
    def __init__(self,brand,speed,color):
        super().__init__(brand,speed)
        self.color=color
    def colorr(self):
        print('Car color is',self.color)

class bike(Vehicle):
    def ring_bell(self):
         print('bike bell rings')

car1=car('bm',60,'black')
car1.colorr()
bik1=bike('mm',80) 
bik1.move()
bik1.ring_bell() 
```

```python
class Animal:
    def __init__(self,name,age,password):
        self.name=name  #public
        self._age=age   #private
        self.__password=password  #protected

    def get_pass(self):
        return self.__password
    
    def set_pass(self,newpassword):
        if len(newpassword)>=6:
            self.__password=newpassword
        else:
            print('password must be at least 6 characters ')
class Dog(Animal):
    def show_data(self):
        print('Public name', self.name)
        print('Protected age', self._age)
        print('Private password', self.get_pass())

dog=Dog('Max',4,'abc123')
dog.show_data()
dog.name='Rex'
dog.show_data()
dog._age=10
dog.show_data()
dog.set_pass('xyz123')
dog.show_data()
```

```python
from abc import ABC, abstractmethod

class Paymment(ABC):
    @abstractmethod
    def pay(self,amount0):
        pass

class CashPayment(Paymment):
    def pay(self,amount):
        print(f'Paid {amount} use cash')

class CardPayment(Paymment):
    def pay(self,amount):
        print(f'Paid {amount} use credit card')

cash=CashPayment()
cash.pay(1000)
card=CardPayment()
card.pay(3000)
```

```python
from abc import ABC,abstractmethod

class vehicle(ABC):
    @abstractmethod
    def move(self,speed):
        pass

class car(vehicle):
    def move(self, speed):
        print( f'car is moving at speed {speed} km/h ')

class plane(vehicle):
    def move(self,speed):
        print(f'Plane is moving at speed {speed} km/h')

car1=car()
car1.move(80)

plane1=plane()
plane1.move(360)
```
