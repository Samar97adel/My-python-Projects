Create a class called `Car`.

It should have:

- `brand`
- `color`

And a method called:

- `show_info()`


````python
class BankAccount:

    def __init__(self,name, balance):
        # private variable (protected from direct access)
        self.__balance = balance
        self.name=name

    # allowed method to deposit money
    def deposit(self, amount):
        self.__balance += amount

    # allowed method to withdraw money
    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
        else:
            print("Not enough balance")

    # allowed method to read balance
    def get_balance(self):
        return self.__balance


# create account
account = BankAccount('Dina',1000)

account.name='ahmed'
account.__balance=100
print(account.name)
# correct way to modify data
account.deposit(500)
print("Balance after deposit:", account.get_balance())

account.withdraw(200)
print("Balance after withdraw:", account.get_balance())


# wrong way: trying to access private variable directly
#print(account.__balance)   # this will raise an error
````

````python

class Car:
    def __init__(self, brand, color):
        self.brand = brand
        self.color = color

    def show_info(self):
        print("Brand:", self.brand)
        print("Color:", self.color)

car1 = Car("Toyota", "Red")
car1.show_info()
````

# 8️⃣ Encapsulation

Encapsulation means we **protect data** and control how it is changed.

Instead of letting anyone change values in any way, we use methods with rules.


## Illustration: Protected Box

<div style="border:2px solid black; border-radius:12px; padding:18px; background:#00000;">
  <div style="font-size:18px; font-weight:bold;">BankAccount</div>
  <div style="margin-top:10px;">💰 balance</div>
  <div style="margin-top:10px;">🔒 Do not change carelessly</div>
  <div style="margin-top:10px;">Use: deposit() / withdraw()</div>
</div>


````python

class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self._balance = balance   # protected by convention

    def deposit(self, amount):
        if amount > 0:
            self._balance += amount
            print("Deposit successful")
        else:
            print("Invalid deposit amount")

    def withdraw(self, amount):
        if amount <= self._balance:
            self._balance -= amount
            print("Withdrawal successful")
        else:
            print("Not enough balance")

    def show_balance(self):
        print("Current balance =", self._balance)

account1 = BankAccount("Omar", 1000)
account1.deposit(300)
account1.withdraw(200)
account1.show_balance()
````

### Why is this better?

Because now we can apply rules such as:

- no negative deposit
- no withdrawal larger than balance

So the object becomes safer and smarter.

# 🔟 Polymorphism

Polymorphism means:

> same method name, different behavior


## Illustration: Same Action, Different Result

````python
class Animal :
    def __init__(self, name ,age):
        self.name=name
        self.age=age 
    def make_sound(self):
        print('Animal is making a sound')

class Dog(Animal): #inheirtance
    def __init__(self,name,age,size):
        super().__init__(name,age)
        self.size=size
    def make_sound(self):   # polymerphism 
        print('dog is barking',self.size)
        
class Cat (Animal):
    def make_sounddd(self):
        print('cat is meowing')

cat=Cat('x',4)
cat.make_sound()
````

## Access Modifiers in Python: Public vs Protected vs Private

In Python, access control is based on naming conventions.

| Type | Syntax | Can be accessed from inside the class? | Can be accessed from child classes? | Can be accessed from outside the class? | Notes |
|---|---|---:|---:|---:|---|
| Public | `name` | ✅ | ✅ | ✅ | Normal attribute or method |
| Protected | `_name` | ✅ | ✅ | ⚠️ Possible, but not recommended | Means: "use inside the class or subclass" |
| Private | `__name` | ✅ | ❌ Direct access is blocked | ❌ Direct access is blocked | Used for stronger hiding through name mangling |

### Quick Comparison

- **Public**: use it when the data or method is safe to access everywhere.
- **Protected**: use it when the data should mainly stay inside the class hierarchy.
- **Private**: use it when you want to hide the internal implementation and control access through methods like getters and setters.

````python
class vehicle:
    def __init__(self,brand,speed):
        self.brand=brand
        self.speed=speed
    def move(self):
        print(self.brand,'is moving at speed',self.speed)

class bike (vehicle):
    def ring_bell(self):
        print('Bike bell ring')

class car (vehicle ):
    def __init__(self,brand,speed,color):
        super().__init__(brand,speed)
        self.color=color
    def colorr(self) :
        print('Car Color is ',self.color)   

              
car1=car('BM',60,'Black')
car1.colorr()
````

````python
class Animal:
    def __init__(self,name,age,password):
        self.name=name  # public
        self._age=age   #protected
        self.__password=password # private

    def get_pass(self):
        return self.__password

    def set_pass(self,new_password):
        if len(new_password)>=6: # length  => number of characters 
            self.__password=new_password
        else:
            print('password must be at least 6 characters  ')    

class Dog (Animal):
    def show_data(self):
        print('Public Name' ,self.name)
        print('Protected Age' ,self._age)
        print('Private Pass' ,self.get_pass())

dog =Dog ('Max',4,'abc123')
dog.show_data()
dog.name='Rex'
dog.show_data()
dog._age=10
dog.show_data()
dog.set_pass('xyz105')
dog.show_data()

````

Even though each object uses the same method name `speak()`, the result is different.

That is the main idea of polymorphism.


# 1️⃣1️⃣ Abstraction

**Abstraction** means we focus on **what an object should do** and hide the unnecessary details of **how it does it**.

In Python, we often use abstraction with **abstract classes**.

An abstract class is like a rule sheet:
- it defines required methods
- child classes must complete them
- we usually do not create objects directly from it


## Real-Life Idea

Think of a remote control.

You press **Power On**.  
You care that the TV turns on.  
You do **not** need to know all the internal electrical steps.

That is abstraction.


````python
from abc import ABC,abstractmethod  # python library  =>زي بستعير كتاب من مكتبة في بايثون 

class Payment(ABC) :
    @abstractmethod
    def pay(self,amount):
        pass 

class CashPayment(Payment):
    def pay(self,amount):
        print(f'paid{amount}using CASH ') 

class CardPayment(Payment) :   
    def pay(self,amount):
        print(f'paid {amount} using Credit Card ') 
        
cash=CashPayment()
cash.pay(1000)

card=CardPayment()
card.pay(500)
````

## Why use abstraction?

- It makes code more organized
- It forces child classes to follow a clear structure
- It is useful when different classes must provide the same method
