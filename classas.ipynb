```python
class Student:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def introduce(self):
        print(f'My name is {self.name}' )
    def birthday(self):
        print(f'My birthday now is {self.age}')
Std1=Student('Samar',29)
Std1.introduce()
Std1.birthday()
```

```python
class account_balance:
    def __init__(self,id,balance):
        self.id=id
        self.__balance=balance
    def deposit(self,amount):
        self.__balance+=amount
    def withdraw(self,amount):
        if self.__balance>=amount:
            self.__balance-=amount
        else:
            print('insufficient')
    def get_balance(self):
        return self.__balance
acc=account_balance(10,15000)
print(acc.id)
acc.get_balance()
acc.__balance=(20000)
acc.get_balance()
acc.deposit(3000)
acc.get_balance()
acc.withdraw(8000)
acc.get_balance()


            
```

```python
class Employee:
    def __init__(self,name,age,salary):
        self.name=name
        self.age=age
        self.__salary=salary
    def deposit(self,amount):
        self.__salary+=amount
    def withdraw(self,amount):
        if self.__salary>=amount:
            self.__salary-=amount
        else :
            print('insufficient')
    def get_salary(self):
        return self.__salary
emp1=Employee('Ahmed', 33, 15000)

print(emp1.name)
print(emp1.age)
emp1.deposit(3000)
emp1.get_salary()
emp1.withdraw(8000)
emp1.get_salary()
```
